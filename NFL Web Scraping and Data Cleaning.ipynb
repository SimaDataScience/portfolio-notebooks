```python
#Import modules.
import time
import selenium
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Function to scrape data from Pro Football Reference from a given starting year to today.
def scrape_nfl_data(startyear, currentyear:int, newestweek:int, save_path:str):
    # Create dictionary of years, weeks.
    previousyears = [i for i in range(startyear, currentyear)]
    fullseason = [i for i in range(1,19)]
    currentseason = [i for i in range(1, newestweek + 1)]
    yearweeks = dict.fromkeys(previousyears, fullseason)
    yearweeks.update(dict.fromkeys([currentyear], currentseason))

    # Initialize list for storing results.
    gameresults = []

    # Scrape pro-football-reference for year, week, team names, and team scores.
    # Loop through each year and week.
    for year, weeks in yearweeks.items():
        for week in weeks:

            # Create link from year and week.
            yearweeklink = 'https://www.pro-football-reference.com/years/{}/week_{}.htm'.format(year, week)

            # Call web driver and get html. If driver crashes, pass and record failure.
            try:
                driver = webdriver.Chrome(ChromeDriverManager().install())
                driver.get(yearweeklink)
                time.sleep(2)
                page = driver.page_source
                driver.quit()
            except:
                page = ""
                
            # Create soup from html.
            soup = BeautifulSoup(page)

            # Filter through each game for the week and pull data. If driver failure, record error.
            for gameblock in soup.find_all('div', class_='game_summary expanded nohover'):
                # Get name and score of losing team.
                losingteam = gameblock.find('tr', class_='loser')
                try:
                    lteamname = losingteam.find('a').text
                    lteamscore = losingteam.find('td', class_='right').text
                    
                except AttributeError:
                    lteamname = 'not found'
                    lteamscore = -1

                #Get name and score of winning team.
                winningteam = gameblock.find('tr', class_='winner')
                try:
                    wteamname = winningteam.find('a').text
                    wteamscore = winningteam.find('td', class_='right').text
                except AttributeError:
                    wteamname = 'not found'
                    wteamscore = -1

                # Append results list.
                game = [year, week, lteamname, lteamscore, wteamname, wteamscore]
                gameresults.append(game)

    # Convert to dataframe and export as csv.
    df = pd.DataFrame(
        gameresults,
        columns=[
            'year',
            'week',
            'losingteamname',
            'losingteamscore',
            'winningteamname',
            'winningteamscore'
        ]
    )
    #df.to_csv(savepath, index=False)
    return df
```

```python
# Initialize variables.
startyear = 2002
currentyear = 2002
newestweek = 18
savepath = '/Users/justinsima/dir/projects/portfolio/eda/NFLDashboard/data/nfl-games.csv'
```

```python
# Scrape data and create dataframe.
df = scrape_nfl_data(startyear, currentyear, newestweek, savepath)
```
